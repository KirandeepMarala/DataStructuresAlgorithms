# Stack Implementation : Arrays

- Linear DataStructure
- Last in First Out

Mainly operations are : 
    1. push()
    2. pop()
    3. peek()
    4. isEmpty()

```python
class Stack:
    
    #initially creating the array [Inbuilt DataStructure in Python]
    def __init__(self):
        self.stack = []
    
    #checking if the array is Empty or not
    def isEmpty(self):
        return self.stack == []
    
    #pushing the element using 'append' operation
    def push(self, ele):
        self.stack.append(ele)
        
    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            return -1
    
    def peek(self):
        if not self.isEmpty():
            return self.stack[-1]
            
        
s = Stack()
while True:
    print("Push")
    print("Pop")
    print('Peek')
    do = input("What exactly you want to do it?")
    if do == 'push':
        ele = int(input("Enter element to insert into the stack: "))
        s.push(ele)
    elif do == 'pop':
        ele = s.pop()
        if ele == -1:
            print("Stack is Empty")
        else:
            print("Deleted element from the stack is: ", ele)
    elif do == 'peek':
        ele = s.peek()
        if ele == -1:
            print("Stack is Empty")
        else:
            print("Top element from the stack is: ", ele)
    else:
        break
            
```

# Stack Implementation : LinkedList

```python
class Node:   
    def __init__(self, info = None):
        self.info = info
        self.link = None
```

```python
class LStack:
    
    #initially creating the stack
    def __init__(self):
        self.head = None
        
    #checking if the array is Empty or not
    def isEmpty(self):
        if self.head == None:
            return True
        else:
            return False
        
    #pushing the element into the stack
    def push(self, ele):
        newNode = Node(ele)
        if self.head == None:
            self.head = newNode
        else:
            newNode.link = self.head
            self.head = newNode
    
    #popping the element
    def pop(self):
        if self.head == None:
            print("No elements left in the List")
        else:
            element = self.head.info
            self.head = self.head.link
            return element
    
    def peek(self):
        if self.head == None:
            print("No elements left in the List")
        else:
            element = self.head.info
            return element
        
        

s = LStack()
while True:
    print("Push")
    print("Pop")
    print('Peek')
    do = input("What exactly you want to do it?")
    if do == 'push':
        ele = int(input("Enter element to insert into the stack: "))
        s.push(ele)
    elif do == 'pop':
        ele = s.pop()
        if ele == -1:
            print("Stack is Empty")
        else:
            print("Deleted element from the stack is: ", ele)
    elif do == 'peek':
        ele = s.peek()
        if ele == -1:
            print("Stack is Empty")
        else:
            print("Top element from the stack is: ", ele)
    else:
        break   
```
