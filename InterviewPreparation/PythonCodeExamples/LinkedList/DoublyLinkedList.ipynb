# DoublyLinkedList

```python
class Node:
    def __init__(self, info, prev = None, next = None):
        self.info = info 
        self.prev = prev
        self.next = next
        
class DLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def insert_at_begining(self, ele):
        newNode = Node(ele)
        if self.head == None:
            self.head = newNode
            self.tail = newNode
        else:
            newNode.next = self.head
            self.head.prev = newNode
            self.head = newNode
            
            
    def insert_at_end(self, ele):
        newNode = Node(ele)
        if self.head == None:
            #If list is empty, both head and tail will point to new node  
            self.head = newNode  
            self.tail = newNode
        else:
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode

    def display(self):
        if self.head == None:
            print("List Empty")
            return 
        current = self.head
        while current != None:
            print(current.info)
            current = current.next
            
    def reverse_display(self):
        current = self.tail
        if self.head == None:
            print("No Elements in the List")
            return 
        if current == self.head:
            print(current.info)
            return 

        while current != self.head:
            print(current.info)
            current = current.prev
            
    def delete_node(self, ele):
        #Base Case:
        if self.head == None:
            print("List is Empty")
            return
        
        #if only one node is present
        if self.head == self.tail:
            if self.head.info == ele:
                temp = self.head
                self.head = None
                temp = None
                return 
            else:
                print("Element is not found in our List")
                return
            
        #delete node in between
        current = self.head.next
        while current.next != None:
            if current.info == ele:
                temp = current.next
                current.prev.next = temp
                temp.prev = current.prev
                temp = None
                return 
            current = current.next
        #delete last node
        if current.info == ele:
            current.prev.next = None
            temp = None
            return 
        print("Element is not Found in the List")
                
        
        
```

```python
ll = DLinkedList()
ll.insert_at_begining(10)
ll.insert_at_begining(100)
ll.insert_at_begining(5)
ll.display()
print("***************")
ll.insert_at_end(20)
ll.insert_at_end(30)
ll.display()
print("***************")
ll.reverse_display()
print("***************")
ll.delete_node(100)
ll.display()
print("***************")
```
