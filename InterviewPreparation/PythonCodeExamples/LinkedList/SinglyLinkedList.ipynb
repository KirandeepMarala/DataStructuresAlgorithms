```python
#For creating the Node we use this Function
class Node:
    #constructor
    def __init__(self, info, link = None):
        self.info = info
        self.link = link
        
#creating the LinkedList class where all operations are done here
class LinkedList:
    
    #at first creating the head by default
    def __init__(self):
        self.head = None
        self.tail = None
    
    #Function used to insert at the Begining
    def insert_at_begining(self, info):
        #creating the node
        newNode = Node(info)
        if self.head != None:
            newNode.link = self.head
            self.head = newNode
        else:
            self.head = newNode
            self.tail = newNode
            
    #Function used to insert at the End
    def insert_at_end(self, info):
        #creating the node
        newNode = Node(info)
        if self.head == None:
            #If list is empty, both head and tail will point to new node  
            self.head = newNode  
            self.tail = newNode
        else:
            #newNode will be added after tail such that tail's next will point to newNode  
            self.tail.link = newNode
            #newNode will become new tail of the list  
            self.tail = newNode

        
       # if self.head != None:
           # current = self.head
            #while current.link != None:
                #current = current.link
            #current.link = newNode
        #else:
           # self.head = newNode  
            
    #Function used to insert at the required Position
    def insert_at_required(self, info, index):
        #creating the node
        if self.head == None:
            self.insert_at_begining(info)
            return       
        newNode = Node(info)
        if index-1 == 0:
            self.insert_at_begining(info)
            return          
        current = self.head  
        count = 0
        while current.link != None:
            count += 1
            if count == (index - 1):
                newNode.link = current.link
                current.link = newNode
                return 
            current = current.link
        self.insert_at_end(info)
    

    #This Function used to delete the info 
    def delete_node(self, ele):
        if self.head == None:
            print("List Empty")
            return 
        if self.head.info == ele:
            temp = self.head
            self.head = temp.link
            temp = None
            return 
        current = self.head
        while current.link != None:
            if current.link.info == ele:
                temp = current.link
                current.link = temp.link
                temp = None
                return
            current = current.link
        print("Element is not in our List")
      
            
    #function used to display all items
    def display(self):
        if self.head == None:
            print("List Empty")
            return 
        current = self.head
        while current != None:
            print(current.info)
            current = current.link
          
    def searching(self, ele):
        if self.head == None:
            print("List is Empty")
            return 
        current = self.head
        while current != None:
            if current.info == ele:
                print("Element is found in list")
                return
            current = current.link
        print("Element Not Found")
```

```python
LL = LinkedList()
LL.insert_at_begining(10)
LL.insert_at_end(20)
LL.insert_at_begining(544)
LL.insert_at_end(60)
LL.insert_at_begining(35)
LL.display()
LL.delete_node(544)
print('*' * 20)
LL.display()
print('*' * 20)
LL.insert_at_required(39, 2)
LL.insert_at_required(22, 3)
LL.insert_at_required(21, 1)
LL.display()
LL.searching(21)
print('*' * 20)
LL.searching(420)
```
